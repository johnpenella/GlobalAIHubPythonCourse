```python
min_ = 0
max_ = 100

def prime(min_,max_):
    primes = []
    for number in range(min_,max_+1):
        if number > 1:
            for i in range(2,number):
                if (number % i) == 0:
                    break
            else:
                primes.append(number)
    return primes

primes = prime(min_,max_)
print('Prime numbers between',min_,'and',max_,'are:')
for prime in primes:
    print(prime)
```
